```python
import numpy as np
```

```python
print("==========Student's Performance==========")

students = np.array([
    [95,65,85],
    [89,72,88],
    [92,75,91]
])

print("\nAll Marks:")
print(students)


print("\nShape:", students.shape)

print('\n=========CLASS STATISTICS=========')

print("\nTotal Marks:", students.sum())
print("Class Average:", students.mean())
print("Highest Marks:", students.max())
print("Lowest Marks:" , students.min())

print("\n==========STUDENT'S PERCENTAGE==========")

student_avegrage = students.mean(axis=1)

print("\nStudent A Percentage:", student_avegrage[0])
print("Student B Percentage:", student_avegrage[1])
print("Student C Percentage:", student_avegrage[2])

print("\n============SUBJECT-WISE AVERAGE============")

subject_average = students.mean(axis = 0)

print("\nSubject A:", subject_average[0])
print("Subject B:", subject_average[1])
print("Subject C:", subject_average[2])

print("\n=======Topper=======")

topper = np.argmax(student_avegrage)

print("\nTopper's Student Index:", topper)
print("Topper's percentage:", student_avegrage[topper])

```
